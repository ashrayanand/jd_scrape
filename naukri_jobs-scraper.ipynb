```
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
from bs4 import BeautifulSoup
import time
import pandas as pd
import re
import requests
import pickle
import json
import time

#driver.get("https://www.naukri.com/mba-jobs?k=mba&ctcFilter=101&ctcFilter=6to10&ctcFilter=10to15&ctcFilter=15to25&ctcFilter=50to75&ctcFilter=75to100&ctcFilter=25to50&jobAge=15&experience=5")
driver.get("https://www.naukri.com/machine-learning-natural-language-processing-deep-learning-computer-vision-data-analyst-data-scientist-data-science-nlp-information-retrieval-jobs?k=machine%20learning%2C%20natural%20language%20processing%2C%20deep%20learning%2C%20computer%20vision%2C%20data%20analyst%2C%20data%20scientist%2C%20data%20science%2C%20nlp%2C%20information%20retrieval")
driver.find_element(By.XPATH,"//i[@class='naukicon naukicon-arrow-1']").click()
driver.find_element(By.XPATH,"//li[contains(text(),'Date')]").click()
time.sleep(5)
driver.execute_script("window.open('about:blank', 'tab2');")
driver.switch_to.window("tab2")
driver.switch_to.window(driver.window_handles[0])
x=1
l=[] 

while x<=500:
    time.sleep(3)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    soup1=soup.find_all("article",{"class":"jobTuple bgWhite br4 mb-8"})
    print("again---",x)   
    for item in soup1:
        #time.sleep(5)
        d={}
        j=item.find("a",{"class":"subTitle ellipsis fleft"}).text
        try:
          d["company_name"]=re.sub("[0-9]","",re.sub("[\(\[].*?[\)\]]","",j))
        except:
          pass
        try:
          d["job_title"]=item.find("a",{"class":"title fw500 ellipsis"}).text
        except:
          pass
        try:
          d["experience"]=item.find("li" ,{"class":"fleft grey-text br2 placeHolderLi experience"}).text
        except:
          pass
        try: 
          d["salary"]=item.find("li",{"class":"fleft grey-text br2 placeHolderLi salary"}).text
        except:
          pass

        try:
          d["location"]=item.find("li",{"class":"fleft grey-text br2 placeHolderLi location"}).text
        except:
          pass
        try:
          d["desc"]=item.find("div",{"class":"job-description fs12 grey-text"}).text
        except:
          pass

        #d["skills"]=item.find("ul",{"class":"tags has-description"}).text  
        
        try:
            details2=item.find("ul",{"class":"tags has-description"})
            details3=details2.find_all("li")
            i=1
            skills=''
            for each2 in details3:
                skills += (str(each2.text) + ' ,')
                i+=1
                if i>8:
                    break
            d["skills"]= skills
        except:
            pass
        
        l.append(d)
            
    x+=1  
    print(x)
    time.sleep(5)
    #driver.switch_to.window(driver.window_handles[0])
    try:
        driver.find_element(By.XPATH,"//div[@class='fleft pages']//a[contains(text(),'"+str(x)+"')]").click()
    except:
        print("done")
        driver.refresh()
        time.sleep(5)
        #break
    #time.sleep(5)
```
